The Boston Housing dataset contains information about various houses in Boston through different parameters. This data was originally a part of UCI Machine Learning Repository and has been removed now.


There are __506 samples and 13 feature variables__ in this dataset. The dataset itself is available at this link. However, because here we are importing this dataset from the scikit-learn itself.

## Exploratory Data Analysis

### Importing Libraries

```python
# Importing the necessary libraries and dataset

# Numerical computing and linear algebra

import numpy as np
import pandas as pd

# Data Visualization

import matplotlib.pyplot as plt
import seaborn as sns

# Importing boston housing dataset

from sklearn.datasets import load_boston

%matplotlib inline
```

### Goal

The objective is to predict the __value of prices of the house__ using the given features. 

```python
# Loading the boston dataset

boston = load_boston()
boston.keys()
```

```python
print(boston.DESCR)
```

```python
# Initializing the features dataframe

features = pd.DataFrame(boston[ "data" ])
features.shape
```

```python
features.columns = boston.feature_names
features.head(10)
```

```python
prices = pd.DataFrame(boston['target'])
print(prices.shape)
prices.columns = ['MEDV']
prices.head(10)

sns.set(rc= {
    "figure.figsize" : (12, 8)
})
sns.scatterplot(data = prices)
```

```python
# Concatenating features and prices to examples.

examples = pd.concat([ features, prices ], axis= 1)
examples.head(10)
```

```python
sns.set(
    rc = {
        "figure.figsize" : (20, 15)
    }
)
plt.hist(
    prices,
    bins=30, 
   color='lightgreen'
)
plt.xlabel("House prices in 1000 USD")

plt.show()
```

```python

sns.heatmap(data=examples.corr().round(2), annot=True)
```

### Model

### Linear Regression

In statistics, __linear regression__ is a linear approach to modelling the relationship between a scalar response and one or more explanatory variables (also known as dependent and independent variables). The case of one explanatory variable is called simple linear regression; for more than one, the process is called __multiple linear regression__. This term is distinct from multivariate linear regression, where multiple correlated dependent variables are predicted, rather than a single scalar variable.

Linear regression is one of the easiest and most popular Machine Learning algorithms. It is a statistical method that is used for predictive analysis. Linear regression makes predictions for continuous/real or numeric variables 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tt_split
from sklearn.metrics import mean_squared_error as mse
```

```python
features_tr, features_te, prices_tr, prices_te= tt_split(   features,
                                                            prices,
                                                            test_size= 0.15,
                                                            random_state= 0 )
```

```python
# Training the model

regressor = LinearRegression()

regressor.fit( features_tr, prices_tr )

price_preds= regressor.predict( features_te )
```

### Prediction Accuracy

```python
accuracy = mse( prices_te , price_preds )

print("Accuracy of the model is ", "%.2f"% accuracy ,"% ." )
```
